# Google Image Scrapper

## Importing Libraries

```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
import logging
import os

```

### Creating Directory for Storing the Images

```python
save_dir='images/'

if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

### If the visiting the Website Again And Again we will be Blocked By Google to Avoid this I'm Creating An Header

```python
query="elon musk"
response = requests.get(f"https://www.google.co.in/search?q={query}&client=safari&sca_esv=562751389&channel=iphone_bm&tbm=isch&sxsrf=AB5stBgilKYxoTugwrH8XThq_ZMy9eDQYQ:1693925502517&source=lnms&sa=X&ved=2ahUKEwj50ume3JOBAxXMUGwGHVZIACkQ_AUoAnoECAcQBA&biw=1440&bih=848&dpr=1#imgrc=j66zmxxPjf2shM")
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
image_tags=soup.find_all("img")
```

```python
image_tags
```

```python
del image_tags[0]
```

```python
len(image_tags)
```

```python
images_data_mongo=[]
for i in image_tags:
    image_url=i['src']
    image_data=requests.get(image_url).content
    mydict={"index":image_url,"image":image_data}
    images_data_mongo.append(mydict)
    with open(os.path.join(save_dir,f"{query}_{image_tags.index(i)}.jpg"),'wb') as f:
        f.write(image_data)
        
    
    
    
    
```
